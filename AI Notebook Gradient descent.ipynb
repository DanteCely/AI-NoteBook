## AI Notebook | Gradient descent

```python
import numpy as np
import scipy as sc

import matplotlib.pyplot as plt
```

```python
# lambda is a anonymous function
# th is theta -> 
# x: th[0] and y: th[1]
func = lambda th: np.sin( (1/2)*(th[0]**2) - (1/4)*(th[1]**2) + 3) * np.cos( (2*th[0]) + 1 - (np.e**th[1]) )

res = 100

_X = np.linspace( -3, 3, res )
_Y = np.linspace( -3, 3, res )

_Z = np.zeros((res, res))

for ix, x in enumerate(_X):
    for iy, y in enumerate(_Y):
        _Z[iy, ix] = func([x,y])
    
plt.contourf(_X, _Y, _Z, 100)
plt.colorbar()
Theta = np.random.rand(2) * 4 - 2 # ratio 0 - 1 to -2 - 2
plt.plot(Theta[0], Theta[1], "o", c="white")

# Evaluar elementos del vector
_T = np.copy(Theta)
h = 0.01
grad = np.zeros(2)
lr = 0.001 # Learning rate

for _ in range(10000):
    for it, th in enumerate(Theta):
        _T = np.copy(Theta)
        _T[it] = _T[it] + h
        deriv = (func(_T) - func(Theta)) / h
        grad[it] = deriv
    Theta = Theta - (lr * grad)
    if( _ % 50 == 0):
        plt.plot(Theta[0], Theta[1], ".", c="red")
    
plt.plot(Theta[0], Theta[1], "o", c="green")

plt.show()
```
